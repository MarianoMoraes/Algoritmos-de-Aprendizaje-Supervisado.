<h3 style=color:blue > Preprocesado de datos para su uso </h3> 

#### Para aplicar algun proceso de tranformacion primero debemos corroborar cual es la distribucion de nuestros datos, para ello utilizaremos el test de normalidad de Shapiro-Wilk en todas las variables para comprobar si siguen una distribucion Gaussiana y saber si dichas variables tienen Outliers, este ultimo paso lo hemos comprobado anteriormente mediante graficos Box Plot.

```python
# Test de normalidad de Shapiro-Wilk para X_test

from scipy.stats import shapiro

for i in df:
    stat, p = shapiro(df[i])
    print('nombre de la variable', i)
    print('stat=%.3f, p=%.3f' % (stat, p))
    
    if p > 0.05:
        print('distribucion gaussiana')
    else:
        print('distribucion NO gaussiana', "\n")
```

####  Mediante el test de Shapiro-Wilk hemos comprobado que ninguna de las variables en ambos dataset siguen una distribucion Gaussiana sabiendo esto asi como tambien las variables que tienen outliers  y que la variable CHAS es binaria ya que solo cuenta con 0 y 1 podemos elegir el modelo Min Max Scaler para las variables sin outliers y Robust Scaler para las variables con outliers.

```python
outliers_variables = ['CRIM', 'ZN','RM', 'DIS', 'PTRATIO', 'B', 'LSTAT'] 
no_outliers_variables = ['INDUS', 'NOX', 'AGE', 'TAX']
```

```python
from sklearn.preprocessing import RobustScaler

scaling_robust = RobustScaler()

df_robust = scaling_robust.fit_transform(df[outliers_variables])

df_robust_scaler = pd.DataFrame(df_robust, columns=outliers_variables)
```

### Linear Regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# linear regression model
linear_model = linear_model.LinearRegression()

# training model with training datta
lm_fit = linear_model.fit(X_train, y_train)

# model prediction
y_pred_lm = linear_model.predict(X_test)

# performance evaluation
mse_lm = mean_squared_error(y_test, y_pred_lm)
r2_lm = r2_score(y_test, y_pred_lm)


# regression coef
coeficientes = lm_fit.coef_
print(f'Coeficientes: {coeficientes}',"\n")

# intercept
intercection = lm_fit.intercept_
print(f'Intercection: {intercection}',"\n")

# Coeficiente de determinacion R2
r2 = lm_fit.score(X_test, y_test)

print(f'R¬≤: {r2}',"\n")

print(f'Mean Squared Error: {mse_lm}',"\n")
```

### Compara el seu rendiment emprant l‚Äôaproximaci√≥ traint/test o emprant totes les dades (validaci√≥ interna).

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Crear instancias de los modelos
linear_regression = LinearRegression()
random_forest = RandomForestRegressor(random_state=42)

# Realizar validaci√≥n cruzada con regresi√≥n lineal
scores_lr = cross_val_score(linear_regression, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
mse_lr_mean = -scores_lr.mean()

# Realizar validaci√≥n cruzada con Random Forest Regressor
scores_rf = cross_val_score(random_forest, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
mse_rf_mean = -scores_rf.mean()

print(f'Mean Squared Error (Regresi√≥n Lineal): {mse_lr_mean}')
print(f'Mean Squared Error (Random Forest): {mse_rf_mean}')
```
